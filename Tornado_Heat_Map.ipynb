```python
import pandas as pd
import folium
from folium.plugins import HeatMap
from IPython.display import display, HTML

df=pd.read_csv("1950-2023_actual_tornadoes.csv")

df=df[df["mag"]>=3].dropna(subset=["slat","slon"])

m=folium.Map(location=[39.5, -98.35], zoom_start=5, tiles="CartoDB dark_matter")

HeatMap(
    df[["slat", "slon"]].values.tolist(),
    radius=15, blur=20,
    gradient={0.2: "back", 0.4: "red", 0.6: "orange", 0.8: "yellow", 1.0: "white"}
).add_to(m)

display(HTML(m._repr_html_()))
```

```python
import pandas as pd

url="https://www.spc.noaa.gov/wcm/data/1950-2023_actual_tornadoes.csv"
df=pd.read_csv(url,low_memory=False)

print(f"Rows loaded:{len(df)}")
print(df.columns.tolist())
```

```python
import folium
from folium.plugins import HeatMap
from IPython.display import display, HTML

df_filtered =df[df["mag"]>=3].dropna(subset=["slat", "slon"])
print(f"Tornado points: {len(df_filtered)}")

m=folium.Map(location=[39.5, -98.35], zoom_start=5,
             tiles="CartoDB dark_matter")

HeatMap(
    df_filtered[["slat", "slon"]].values.tolist(),
    radius=15, blur=20, min_opacity=0.4,
    gradient={0.2: "black", 0.4: "red",
              0.6: "orange", 0.8: "yellow", 1.0: "white"}
).add_to(m)

display(HTML(m._repr_html_()))
```
